### Tenserflow & Keras
Tensorflow Google tarafından geliştirilen çok kapsamlı alanlarda kullanılan kütüphanelerden biridir. Tensorflow Keras ile birlikte kullanılabilmesi sayesinde anlaşılması ve kullanılması kolay bir hale geldi. Artık Keras Tensorflow içerisinde API olarak çalışabiliyor.  
  
Bu notebookta bilinen *"iris"* veri setini kullanarak Tensorflow ve Keras kullanarak derin öğrenme analizinin nasıl yapıldığını görebileceğiz.

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris.keys() # .keys() ile iris 'in neler barındırdığını görebiliriz.
```

`DESCR` anahtarı ile irisin özet verilerini görebiliriz.

```python
print(iris['DESCR'])
```

```python
import pandas as pd
import numpy as np 

X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.DataFrame(iris.target, columns=['Type'])

df = pd.concat([X, y], axis=1)
df.to_csv('../Datasets/iris_data.csv', index=False)

df.head()
```

```python
df['Type'].value_counts()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline

sns.pairplot(X)
plt.savefig('../Images/pairplot.png')
plt.show()
```
